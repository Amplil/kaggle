```python
# show the relationshaoe between Pclass and Survived

from icecream import ic
import  matplotlib.pyplot as plt
survived_Pclass = data_train['Pclass'].groupby(data_train['Survived'])
ic(survived_Pclass.value_counts().unstack())
survived_Pclass.value_counts().unstack().plot(kind='bar', stacked='True')
plt.show()
```

```python
# show the relationshaoe between Sex and Survived

survived_Sex = data_train['Sex'].groupby(data_train['Survived'])
# survived_Sex.value_counts().unstack()
survived_Sex.value_counts().unstack().plot(kind='bar', stacked='True')
plt.show()
```

```python
# show the relationshaoe between Age and Survived
# filling missing values by fillna()

data_train['Age'] = data_train['Age'].fillna(data_train['Age'].mean())
# ic(data_train['Age'].isnull().sum()) # 0

age_data_train = pd.cut(data_train['Age'], bins=[0, 5, 10, 15, 25, 40, 65, 100])

# survived_Age = pd.DataFrame(data_train['Survived'].values, age_data_train, columns=['Survived', ])
# ic(survived_Age.head())
ct = ic(pd.crosstab(age_data_train, data_train['Survived'].astype(int))) 
# ct.plot(kind='bar', stacked=True)
ct.plot()
plt.show()
```

```python
# filling missing values by fillna()
data_train['Embarked'] = data_train['Embarked'].fillna(ic(data_train['Embarked'].value_counts().idxmax()))
```

```python
# test data preprocess
data_test.info()
```

```python
# fill in the missing data age
data_test['Age'] = data_test['Age'].fillna(data_test['Age'].mean())
```

```python
# man for 0 and fimale for 1
data_train.loc[data_train['Sex']=='male', "Sex"] = 0
data_train.loc[data_train['Sex']=='female', "Sex"] = 1
data_test.loc[data_test['Sex']=='male', "Sex"] = 0
data_test.loc[data_test['Sex']=='female', "Sex"] = 1

# S for 0, C for 1, Q for 2
data_train.loc[data_train['Embarked']=='S', "Embarked"] = 0
data_train.loc[data_train['Embarked']=='C', "Embarked"] = 1
data_train.loc[data_train['Embarked']=='Q', "Embarked"] = 2
data_test.loc[data_test['Embarked']=='S', "Embarked"] = 0
data_test.loc[data_test['Embarked']=='C', "Embarked"] = 1
data_test.loc[data_test['Embarked']=='Q', "Embarked"] = 2

"""# normlization unnesscery"""
```

```python
# drop the useless columns
dropList = ['PassengerId', 'Name', 'Ticket', 'Fare', 'Cabin']
if 'PassengerId' in data_train:
    data_train = data_train.drop(dropList,axis=1)
    data_test = data_test.drop(dropList,axis=1)

ic(data_train.info())
ic(data_test.info())
```

```python
from sklearn.tree import DecisionTreeClassifier, export_graphviz
import graphviz
import pydotplus
from IPython.display import Image
# from sklearn.externals.six import StringIO
from six import StringIO

# data
X = data_train[['Pclass','Sex', 'Age', 'SibSp', 'Parch', 'Embarked']]
Y = data_train[['Survived']]

clf = DecisionTreeClassifier()
clf = clf.fit(X, Y)
clf.classes_
```

```python
dot_data = StringIO() #dotファイル情報の格納先
export_graphviz(clf, out_file=dot_data,  
                     feature_names=['Pclass','Sex', 'Age', 'SibSp', 'Parch', 'Embarked'],
                     class_names=['Not survived','Survived'],
                     filled=True, rounded=True,  
                     special_characters=True)
graph = pydotplus.graph_from_dot_data(dot_data.getvalue()) 
Image(graph.create_png())
```

```python
clf.score(X,Y)
```

```python
# test
X_test = data_test[['Pclass','Sex', 'Age', 'SibSp', 'Parch', 'Embarked']]
y_pred = clf.predict(X_test)
```
